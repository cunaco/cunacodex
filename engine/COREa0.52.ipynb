**ИМПОРТ БИБЛИОТЕК**

```python
import random
import numpy as np
import spacy
import flask
import tkinter
import tensorflow
import keras

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

**ИМПОРТ СЛОВАРЕЙ ВОПРОСОВ CELEST1A И ОТВЕТОВ C2GB**

```python
CELEST1A = {
    'MIND': ['CAPACITY', 'COMPUTING', 'DIRECTION', 'REFLECTION'],
    'BODY': ['STRUCTURE', 'APPEARENCE', 'COMPLEXITY', 'STRENGTH'],
    'SOUL': ['IMAGE', 'MORAL', 'PASSION','WEIGHT'],
    'WILL': ['CUT','GO','PICK','SPACE']
}

C2GB = {
    'RELIGION': ['PERCEPTION', 'CONCEPTION', 'ACCEPTION', 'RECEPTION'],
    'PHILOSOPHY': ['RATING', 'TOPICAL', 'TERMINAL', 'COMPLEX'],
    'ART': ['PAST', 'PRESENT', 'FUTURE', 'ETERNAL'],
    'SCIENCE': ['CROWN', 'FRUIT', 'TRUNK', 'ROOTS']
}
```

**ИМПОРТ СЛОВАРЯ МАРШРУТОВ N4M7C**

```python
N4M7C = {
     'C1': ['cancellation'],
     'C2': ['control'],
     'C3': ['cinematics'],
     'C4': ['configurative'],
     'C5': ['care'],
     'С6': ['capable'],
     'С7': ['cheer']
}
```

**ИМПОРТ СЛОВАРЯ ПРИНЦИПОВ R4ST**


```python
R4ST = {
    'FEED': ['INFO','FOOD','MANA','GRACE'],
    'STIMULUS': ['COMMUNICATION','MASTER','ARCADE','FREEPLAY'],
    'PULSE': ['HIGH','MIDDLE','LOW','OUT'],
    'SYNC': ['EXPERIENCE','PRODUCT','RELATION','SOLVERY']
}
```

**ПРЕОБРАЗОВАНИЕ СЛОВАРЕЙ ОПЕРАЦИЙ И АТРИБУТОВ ИЗ СЛОВ В ЧИСЛА**

```python
CELEST1A_numeric = np.array([list(MIND) for MIND in CELEST1A.values()]);
CELEST1A_numeric = np.array([list(BODY) for BODY in CELEST1A.values()]);
CELEST1A_numeric = np.array([list(SOUL) for SOUL in CELEST1A.values()]);
CELEST1A_numeric = np.array([list(WILL) for WILL in CELEST1A.values()]);

C2GB_numeric = np.array([list(RELIGION) for RELIGION in C2GB.values()]);
C2GB_numeric = np.array([list(PHILOSOPHY) for PHILOSOPHY in C2GB.values()]);
C2GB_numeric = np.array([list(ART) for ART in C2GB.values()]);
C2GB_numeric = np.array([list(SCIENCE) for SCIENCE in C2GB.values()]);

N4M7C_numeric = np.array([list(C1) for C1 in N4M7C.values()]);
N4M7C_numeric = np.array([list(C2) for C2 in N4M7C.values()]);
N4M7C_numeric = np.array([list(C3) for C3 in N4M7C.values()]);
N4M7C_numeric = np.array([list(C4) for C4 in N4M7C.values()]);
N4M7C_numeric = np.array([list(C5) for C5 in N4M7C.values()]);
N4M7C_numeric = np.array([list(C6) for C6 in N4M7C.values()]);
N4M7C_numeric = np.array([list(C7) for C7 in N4M7C.values()]);

R4ST_numeric = np.array([list(FEED) for FEED in R4ST.values()]);
R4ST_numeric = np.array([list(STIMULUS) for STIMULUS in R4ST.values()]);
R4ST_numeric = np.array([list(PULSE) for PULSE in R4ST.values()]);
R4ST_numeric = np.array([list(SYNC) for SYNC in R4ST.values()]);
```

**ПОДГОТОВКА ДАННЫХ ДЛЯ ОБУЧЕНИЯ**

```python
all_texts = []
for category in [CELEST1A, C2GB, N4M7C, R4ST]:
    for subcategory in category.values():
        all_texts.extend(subcategory)

tokenizer = Tokenizer()
tokenizer.fit_on_texts(all_texts)

sequences = tokenizer.texts_to_sequences(all_texts)
max_sequence_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=12, padding='post')
```

**ПРЕОБРАЗОВАНИЕ ДАННЫХ ИЗ СЛОВАРЕЙ В СПИСОК СТРОК**

```python
input_CELEST1A = []
for MIND in CELEST1A:
    input_CELEST1A.extend(CELEST1A[MIND])
for BODY in CELEST1A:
    input_CELEST1A.extend(CELEST1A[BODY])
for SOUL in CELEST1A:
    input_CELEST1A.extend(CELEST1A[SOUL])
for WILL in CELEST1A:
    input_CELEST1A.extend(CELEST1A[WILL])

input_C2GB = []
for RELIGION in C2GB:
    input_C2GB.extend(C2GB[RELIGION])
for PHILOSOPHY in C2GB:
    input_C2GB.extend(C2GB[PHILOSOPHY])
for ART in C2GB:
    input_C2GB.extend(C2GB[ART])
for SCIENCE in C2GB:
    input_C2GB.extend(C2GB[SCIENCE])

input_N4M7C = []
for FEED in N4M7C:
    input_N4M7C.extend(N4M7C[FEED])
for STIMULUS in N4M7C:
    input_N4M7C.extend(N4M7C[STIMULUS])
for PULSE in N4M7C:
    input_N4M7C.extend(N4M7C[PULSE])
for SYNTHESIS in N4M7C:
    input_N4M7C.extend(N4M7C[SYNTHESIS])

input_R4ST = []
for ABLE in R4ST:
    input_R4ST.extend(R4ST[ABLE])
for CINEMATIC in R4ST:
    input_R4ST.extend(R4ST[CINEMATIC])
for CONFIGURATIVE in R4ST:
    input_R4ST.extend(R4ST[CONFIGURATIVE])
for CHEERING in R4ST:
    input_R4ST.extend(R4ST[CHEERING])
```

**СОЗДАНИЕ И ОБУЧЕНИЕ НЕЙРОСЕТИ**

```python
# Определяем размер словаря (количество уникальных слов)
vocab_size = len(tokenizer.word_index) + 1

# Создаем модель нейронной сети
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=100, input_length=max_sequence_length))
model.add(LSTM(units=128))
model.add(Dense(units=vocab_size, activation='softmax'))

# Компилируем модель
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Обучаем модель
model.summary()
```

**ИНТЕРФЕЙС**

```python
while True:
    user_input = input("Вы: ")
    user_input_sequence = tokenizer.texts_to_sequences([user_input])
    user_input_sequence_padded = pad_sequences(user_input_sequence, maxlen=max_sequence_length, padding='post')
    predicted_probabilities = model.predict(user_input_sequence_padded)[0]
    predicted_index = np.argmax(predicted_probabilities)

    # Определение категории по индексу
    while predicted_index < len(CELEST1A):
        predicted_response = list(CELEST1A.values())[predicted_index]
        predicted_response = list(C2GB.values())[predicted_index]
        predicted_response = list(N4M7C.values())[predicted_index]
        predicted_response = list(R4ST.values())[predicted_index]
    print("VARRAWA:", predicted_index)
while False:
    print("VARRAWA: Я не могу тебе помочь!")
    break
```
